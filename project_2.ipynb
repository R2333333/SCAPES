##Encoding Class Labels

```
#assign subreddit to a number id, return dictionary key = number, value = subreddit name
def createDictionary(classLabelArray):
  length = len(classLabelArray)
  numCategories = 0
  dict = {}
  for i in range(length):
    if i==0:
      dict[numCategories] = classLabelArray[i]
    else:
      if not classLabelArray[i] in dict.values():
        numCategories = numCategories+1
        dict[numCategories] = classLabelArray[i]
  return dict
```

```
#encode y column values replacing name of subreddit with assigned id number
def convertClassLabelStrings(classLabelArray):
  dictionary = createDictionary(classLabelArray)
  length = len(classLabelArray)
  encoded = []

  for i in range(length):
    search = classLabelArray[i]
    for key, value in dictionary.items():
      if value == search:
        encoded.append(key)
  return encoded
```
